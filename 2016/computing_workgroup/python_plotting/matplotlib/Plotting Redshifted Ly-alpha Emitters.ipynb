```python
import numpy as np  
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
import pylab
from matplotlib.patches import Rectangle

%matplotlib inline
```

```python
#J1 Filter:
#equations for the four wavelengths wrt the J1 filter
x = np.linspace(0,16,100)
y1 = -2.5*np.log10(1+((50*(1+x))/1030))
y2 = -2.5*np.log10(1+((100*(1+x))/1030))
y3 = -2.5*np.log10(1+((200*(1+x))/1030))
y4 = -2.5*np.log10(1+((500*(1+x))/1030))

#plots the four wavelengths wrt the J1 filter
fig, axes = plt.subplots(5, 1, sharex=True, sharey=False, figsize=(15,15))
plt.suptitle(r'$\Delta$(mag) vs redshift', fontsize=14, fontweight='bold')

axes[0].plot(x,y1, label='50ang', color='black')
axes[0].plot(x,y2, label='100ang', color='black')
axes[0].plot(x,y3, label='200ang', color='black')
axes[0].plot(x,y4, label='500ang', color='black')

#plots the Ly-alpha box
axes[0].add_patch(Rectangle((7.211, 0.0), 0.847, -2.5, facecolor='#005c99', alpha=0.9))

#labelling the subplot and individual lines
axes[0].set_ylabel('J1 Filter', fontweight='bold')
#axes[0].text(7.64, -2.0, 'Ly-$\alpha$', rotation=90, verticalalignment='center')
axes[0].text(15.75, -0.6, '50$\AA$', horizontalalignment='right')
axes[0].text(15.75, -1.0, '100$\AA$', horizontalalignment='right')
axes[0].text(15.75, -1.5, '200$\AA$', horizontalalignment='right')
axes[0].text(15.75, -2.3, '500$\AA$', horizontalalignment='right')
axes[0].text(7.64, -1.2, 'Ly-'+r'$\alpha$',
        horizontalalignment='center',
        verticalalignment='center',
        rotation='vertical')
```
